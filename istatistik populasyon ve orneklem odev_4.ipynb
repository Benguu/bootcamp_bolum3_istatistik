```python
#soru_1
#iki normal dağılıma sahip rassal sayı üretim. 
#Bu rassal sayılardan birisinden 1000 tane 10.000 tane elde edelim. 
#Bu iki rassal sayının da olasılık yoğunluk grafiğini çizelim. Sizce Merkezi Limit Theoremi bu olayda geçerli midir?

import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.stats import norm
from scipy.stats import ttest_ind

rand1 = np.random.normal(0, 2, 1000)
rand2 = np.random.normal(0,1, 10000) 
varyans=2
sigma = math.sqrt(varyans)
x = np.linspace(0 - 2*sigma, 0 + 2*sigma, 100)
plt.plot(x,norm.pdf(x, 0, sigma))
print(ttest_ind(rand1, rand2, equal_var=False))
# Her iki grup için histogram çizimi

plt.hist(rand1, label='Population 1') 
plt.hist(rand2, label='Population 2') 
plt.legend(loc='upper right') 
plt.show()


```

```python
#soru_2
#İki madeni paramızın olduğunu varsayalım. 
#Bunlardan bir tanesi adil (yani, yeteri kadar deneme yapıldığında yazı veya tura 1/2 olasılığına yaklaşmakta) diğeri ile ise 3/4 olasılığında tura elde edilmektedir.
#Herhangi biri seçilip, n kere deneme yapıldığında,Sn tura sayısı olursa, Büyük Sayılar Yasası uzun vadede gelecek tura sayısını tahmin etmemize izin verir mi? 
#Çok sayıda deneme yapıldığında, hangi paranın seçildiğini bilebilir miyiz? Kaç deneme bizi %95 oranında emin kılar?
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt 
np.random.seed(42)
n = 10
p1 = 0.5
p2 = 0.75
size=100
para=np.random.binomial(n,p1, size=size) #1/2 olasılıkla tura elde etme
hpara=np.random.binomial(n,p2,size=size) #3/4 olasılıkla tura elde etme sayısı
olspara =pd.DataFrame([np.equal(para,i).mean() for i in range(n)]) #normal para tura elde etme olasılığı
olshpara = pd.DataFrame([np.equal(hpara,i).mean() for i in range(n)])# adil olmayan para tura elde etme olasılığı
plt.plot(list(range(n)), olspara, color='blue', marker='o')
plt.xlabel('Tura Sayısı',fontsize=14)
plt.ylabel('Olasılık',fontsize=14)
plt.plot(list(range(n)), olshpara, color='red', marker='o')
plt.xlabel('Hatalı Para Tura Sayısı',fontsize=14)
plt.ylabel('Olasılık',fontsize=14)

# Grafikte normal para için tura sayısı 0,5 olasılıklı adil olmayan para için tura sayısı olasılığının 0,7  olduğu gözlemlenmiştir.
```
